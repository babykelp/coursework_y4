```python
import pandas as pd
from matplotlib import pyplot as plt
import wesanderson
import numpy as np
import seaborn as sns
from labellines import labelLines
```

```python
data = pd.read_csv('thermal_conduct - Sheet1.csv')

moss = data[data.type == 'Moss']
sand = data[data.type == 'Sand']
clay = data[data.type == 'Clay']
peat= data[data.type == 'Peat']

color = wesanderson.film_palette("Moonrise Kingdom",1)
```

```python
fig, axes = plt.subplots(1,2,figsize =(8,4))

axes[0].plot(sand.water_content, sand.soil_water_TC,label = '', color= color[1], linewidth = 2)
axes[0].plot(clay.water_content, clay.soil_water_TC,label = '', color= color[2], linewidth = 2)
axes[0].grid()
axes[0].set_xlabel('    [/]', fontsize=12)
axes[0].set_ylabel(' [/*]', fontsize=12)
axes[0].set_xticks(np.arange(0, 1.1, 0.1))

axes[1].plot(peat.water_content, peat.soil_water_TC,label = '', color= color[3], linewidth = 2)
axes[1].plot(moss.water_content, moss.soil_water_TC, label = '', color= color[0], linewidth = 2)
axes[1].grid()
axes[1].set_xlabel('    [/]', fontsize=12)
axes[1].set_xticks(np.arange(0, 1.1, 0.1))

fig.legend(loc='lower center', fontsize=10, bbox_to_anchor=(0.5, -0.05), ncol=4, fancybox=True, shadow=True)
fig.tight_layout(pad = 2)
```

```python
fig, axes = plt.subplots(1,2,figsize =(8,4))

axes[0].plot(sand.ice_content, sand.soil_ice_TC,label = '', color= color[1], linewidth = 2)
axes[0].plot(clay.ice_content, clay.soil_ice_TC,label = '', color= color[2], linewidth = 2)
axes[0].grid()
axes[0].set_xlabel('    [/]', fontsize=12)
axes[0].set_ylabel(' [/*]', fontsize=12)
axes[0].set_xticks(np.arange(0, 1.1, 0.1))

axes[1].plot(peat.ice_content, peat.soil_ice_TC,label = '', color= color[3], linewidth = 2)
axes[1].plot(moss.ice_content, moss.soil_ice_TC, label = '', color= color[0], linewidth = 2)
axes[1].grid()
axes[1].set_xlabel('    [/]', fontsize=12)
axes[1].set_xticks(np.arange(0, 1.1, 0.1))

fig.legend(loc='lower center', fontsize=10, bbox_to_anchor=(0.5, -0.05), ncol=4, fancybox=True, shadow=True)
fig.tight_layout(pad = 2)
```

```python
data = pd.read_csv('thermal_conduct - rmse.csv')
tests = np.arange(1,12)
df_tests = pd.DataFrame(index = data.depth.unique())
for i in tests:
    df_tests[i] = data[data['test '] == i].set_index(data.depth.unique())['rmse_mean']
display(df_tests)
```

```python
#    (  )
colors = [
    '#b0b0b0',  # 1
    '#b0b0b0',  # 2
    "#b0b0b0",  # 3
    '#b0b0b0',  # 4
    '#b0b0b0',  # 5
    '#b0b0b0',  # 6
    '#e61f3dff',  # 7
    "#b0b0b0",  # 8
    "#b0b0b0",  # 9
    "#47a0a0ff",  # 10
    "#b0b0b0"   # 11
]

fig, ax = plt.subplots(figsize =(8,6))
for i, column in enumerate(df_tests.columns):
    if column in [7, 10]:
        ax.plot(df_tests.index, df_tests[column], label=column, linewidth=2, color=colors[i])
    else:
        ax.plot(df_tests.index, df_tests[column], label=column, linewidth=2, color=colors[i], alpha = 0.5)  #  

plt.grid()
plt.xlabel('  ', fontsize=12)
plt.ylabel('RMSE', fontsize=12)

labelLines(ax.get_lines(), zorder=2.5)

fig.tight_layout(pad=2)
plt.show()
```
